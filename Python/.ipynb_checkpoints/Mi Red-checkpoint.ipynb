```python
data = pd.read_csv('completo+respuesta+nombres.csv', header=0, index_col=False)
```

```python
data.describe()
```

```python
columns=['X-axis', 'Y-axis', 'Z-axis', 'Label']
```

```python
data['X-axis'].mean()
```

```python
x_st=(data['X-axis']-data['X-axis'].mean())/data['X-axis'].std()
y_st=(data['Y-axis']-data['Y-axis'].mean())/data['Y-axis'].std()
z_st=(data['Z-axis']-data['Z-axis'].mean())/data['Z-axis'].std()
```

```python
data_st = pd.concat([x_st, y_st, z_st, data['Label']], axis=1, keys=columns)
```

```python
print(data_st.dtypes)
```

```python
data_st
```

```python
data_st.describe()
```

```python
x_axis = data_st['X-axis']
y_axis = data_st['Y-axis']
z_axis = data_st['Z-axis']
y_label = data_st['Label']
datos_train = pd.concat([x_axis, y_axis, z_axis, y_label], axis=1, keys=['X-axis_st', 'Y-axis_st', 'Z-axis_st', 'Label_st'])
```

```python
categorical = tf.feature_column.categorical_column_with_vocabulary_list('Label_st', ['Reposo', 'Andando', 'Corriendo', 'Cayendo'], default_value=0)

x_axis_column = tf.feature_column.numeric_column('X-axis_st')
y_axis_column = tf.feature_column.numeric_column('Y-axis_st')
z_axis_column = tf.feature_column.numeric_column('Z-axis_st')
label_axis_column = tf.feature_column.embedding_column(categorical_column=categorical, dimension=4)
x_data = datos_train.drop('Label_st', axis=1)
y_data = datos_train['Label_st']
```

```python
feat_cols = [x_axis_column, y_axis_column, z_axis_column, label_axis_column]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_data, 
                                                 y=y_data, 
                                                 batch_size=10, 
                                                 num_epochs=10, 
                                                 shuffle=True)
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[100, 100, 4], 
                                       feature_columns=feat_cols, 
                                       n_classes=4)
```

```python
dnn_model.train(input_fn=input_func, steps = 10)
```
